```python
# Colab 셀 3 (수정 + 확인 코드 추가)
print(f"단계 3: {url} 에서 웹 페이지 내용 가져오기 시도...")

# 이전에 정의된 html_content 초기화 (혹시 모를 이전 실행 영향 제거)
html_content = ""
response = None

try:
    # timeout 값을 넉넉하게 20초로 설정
    response = requests.get(url, headers=headers, timeout=20)
    response.raise_for_status()

    print(f"요청 성공! 상태 코드: {response.status_code}")
    html_content = response.text
    print("HTML 내용 가져오기 완료.")

    # --- 확인 코드 추가 ---
    print("-" * 20)
    print("확인: 가져온 HTML 내용 검사")
    if html_content:
        print(f"HTML content 길이: {len(html_content)}")
        print(f"HTML content 시작 부분 (500자):\n{html_content[:500]}")
    else:
        print("HTML content가 비어있습니다!")
    print("-" * 20)
    # --- 확인 코드 끝 ---

except requests.exceptions.RequestException as e:
    print(f"오류 발생: 웹 페이지 내용을 가져오는 데 실패했습니다.")
    print(f"에러 상세 정보: {e}")
except Exception as e:
    print(f"알 수 없는 오류 발생: {e}")
```

```python
# Colab 셀 2: 라이브러리 불러오기 및 WebDriver 옵션 설정
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from bs4 import BeautifulSoup

print("단계 2: 라이브러리 불러오기 완료.")

# Chrome WebDriver 옵션 설정 (Colab 환경용)
options = webdriver.ChromeOptions()
options.add_argument('--headless') # 브라우저 창 표시 안 함
options.add_argument('--no-sandbox') # Colab에서 필수
options.add_argument('--disable-dev-shm-usage') # Colab에서 메모리 관련 문제 방지
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36") # User-Agent 설정

print("WebDriver 옵션 설정 완료.")
```

```python
# Colab 셀 3: WebDriver 시작 및 페이지 로드
print("단계 3: WebDriver 시작 및 페이지 로드 시도...")

# WebDriver 인스턴스 생성 (설정한 옵션 적용)
driver = webdriver.Chrome(options=options) # Colab에서는 'chromedriver' 경로 자동 인식

# 대상 URL 설정
url = "https://product.kyobobook.co.kr/detail/S000000620181"

try:
    # URL 열기
    driver.get(url)
    print(f"페이지 로드 요청 완료: {url}")
    # 페이지 로딩 및 초기 스크립트 실행 위해 잠시 대기 (2초)
    time.sleep(2)
    print(f"현재 페이지 제목: {driver.title}") # 페이지가 제대로 로드되었는지 제목으로 확인
except Exception as e:
    print(f"WebDriver 시작 또는 페이지 로드 중 오류 발생: {e}")
    driver.quit() # 오류 발생 시 드라이버 종료
    driver = None # 드라이버 변수 초기화
```

```python
# Colab 셀 4: 리뷰 요소 로드 대기
# 이전 단계에서 driver가 성공적으로 생성되었는지 확인
if driver:
    print("단계 4: 리뷰 요소(comment_text)가 나타날 때까지 대기 (최대 15초)...")
    try:
        # WebDriverWait 설정 (driver, 최대 대기 시간(초))
        wait = WebDriverWait(driver, 15)

        # 기다릴 조건 설정: 'comment_text' 클래스를 가진 요소 중 *하나라도* 나타날 때까지
        # (By.CLASS_NAME 사용)
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'comment_text')))

        print("리뷰 요소 로드 확인됨.")

        # --- 추가: "더보기" 버튼 처리 (필요한 경우) ---
        # 교보문고 리뷰에 '더보기' 버튼이 있는지 확인하고, 있다면 모두 로드될 때까지 클릭
        # (현재 구조에서는 더보기 버튼이 바로 보이지 않으므로 일단 주석 처리)
        # while True:
        #     try:
        #         # '더보기' 버튼 찾기 (예: 클래스 이름이 'btn_more_comment' 라고 가정)
        #         load_more_button = driver.find_element(By.CLASS_NAME, 'btn_more_comment')
        #         if load_more_button.is_displayed() and load_more_button.is_enabled():
        #             print("리뷰 '더보기' 버튼 클릭...")
        #             load_more_button.click()
        #             time.sleep(2) # 새 리뷰 로드 대기
        #         else:
        #             print("'더보기' 버튼이 더 이상 표시되지 않거나 비활성화됨.")
        #             break
        #     except NoSuchElementException:
        #         print("'더보기' 버튼을 찾을 수 없음 (모든 리뷰가 로드되었거나 버튼 없음).")
        #         break
        #     except Exception as e:
        #         print(f"'더보기' 처리 중 오류: {e}")
        #         break
        # --- '더보기' 처리 끝 ---

    except TimeoutException:
        print("오류: 최대 대기 시간(15초) 동안 'comment_text' 요소를 찾지 못했습니다.")
        print("페이지 로딩이 매우 느리거나, 요소 클래스 이름이 다르거나, 리뷰가 없을 수 있습니다.")
        # 리뷰가 없어도 오류는 아니므로 드라이버를 종료하지는 않음
    except Exception as e:
        print(f"요소 대기 중 예상치 못한 오류 발생: {e}")
        driver.quit()
        driver = None
else:
    print("단계 4: 이전 단계에서 WebDriver가 제대로 시작되지 않아 건너<0xEB>니다.")
```

```python
# Colab 셀 5: 최종 페이지 소스 가져오기, WebDriver 종료, 파싱
page_source = ""
soup = None

# 이전 단계에서 driver가 유효한지 확인
if driver:
    print("단계 5: JavaScript 실행 후 최종 페이지 소스 가져오기...")
    page_source = driver.page_source
    print(f"페이지 소스 가져오기 완료 (길이: {len(page_source)}).")

    print("WebDriver 종료 중...")
    driver.quit()
    print("WebDriver 종료 완료.")

    # 페이지 소스를 BeautifulSoup로 파싱
    if page_source:
        print("가져온 페이지 소스를 BeautifulSoup로 파싱 시작...")
        soup = BeautifulSoup(page_source, 'html.parser')
        print("파싱 완료.")
    else:
        print("페이지 소스가 비어있어 파싱할 수 없습니다.")
else:
    print("단계 5: WebDriver가 없어 페이지 소스 가져오기 및 파싱을 건너<0xEB>니다.")

# 단계 5 실행 후 page_source 변수 내용 확인 가능
# 예: print(page_source[:1000])
```

```python
# Colab 셀 6 (수정)
extracted_reviews = [] # 결과 저장 리스트 초기화

# 이전 단계에서 soup 객체가 성공적으로 생성되었는지 확인
if soup:
    # print("단계 6: 파싱된 HTML에서 'comment_text' 클래스를 가진 <p> 태그 찾고 텍스트 추출 시도...") # 이전 코드 주석 처리
    print("단계 6: 파싱된 HTML에서 'comment_text' 클래스를 가진 *모든* 태그 찾고 텍스트 추출 시도...") # 수정된 설명

    # soup.find_all('p', class_='comment_text') 대신 아래 코드로 변경
    # 태그 종류('p')를 지정하지 않고 클래스 이름으로만 검색
    review_elements = soup.find_all(class_='comment_text') # <---- 이 부분이 변경됨

    if review_elements:
        print(f"총 {len(review_elements)}개의 'comment_text' 클래스 요소를 찾았습니다.") # 메시지 약간 수정
        for index, element in enumerate(review_elements):
            # 요소의 태그 이름도 함께 출력해 봅시다 (디버깅에 도움됨)
            tag_name = element.name
            review_text = element.get_text(strip=True)
            if review_text:
                print(f"  - 요소 {index+1} (태그: <{tag_name}>) 에서 텍스트 추출 성공")
                extracted_reviews.append(review_text)
            # else:
            #    print(f"  - 요소 {index+1} (태그: <{tag_name}>) 는 텍스트 내용이 비어있습니다.")

        print(f"텍스트 추출 완료. 총 {len(extracted_reviews)}개의 비어있지 않은 리뷰 텍스트 추출.")
    else:
        # 메시지 약간 수정
        print("실패: 파싱된 HTML에서 'comment_text' 클래스를 가진 요소를 찾을 수 없습니다.")
        print("    페이지 구조가 예상과 다르거나, 해당 클래스의 요소가 없을 수 있습니다.")
else:
    print("단계 6: soup 객체가 없어 텍스트 추출을 건너<0xEB>니다.")
```
