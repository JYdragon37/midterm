```python
# Colab  2:    WebDriver  
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from bs4 import BeautifulSoup

print(" 2:   .")

# Chrome WebDriver   (Colab )
options = webdriver.ChromeOptions()
options.add_argument('--headless') #     
options.add_argument('--no-sandbox') # Colab 
options.add_argument('--disable-dev-shm-usage') # Colab    
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36") # User-Agent 

print("WebDriver   .")
```

```python
# Colab  3: WebDriver    
print(" 3: WebDriver     ...")

# WebDriver   (  )
driver = webdriver.Chrome(options=options) # Colab 'chromedriver'   

#  URL 
url = "https://product.kyobobook.co.kr/detail/S000000620181"

try:
    # URL 
    driver.get(url)
    print(f"   : {url}")
    #          (2)
    time.sleep(2)
    print(f"  : {driver.title}") #     
except Exception as e:
    print(f"WebDriver       : {e}")
    driver.quit() #     
    driver = None #   
```

```python
# Colab  5:    , WebDriver , 
page_source = ""
soup = None

#   driver  
if driver:
    print(" 5: JavaScript      ...")
    page_source = driver.page_source
    print(f"    (: {len(page_source)}).")

    print("WebDriver  ...")
    driver.quit()
    print("WebDriver  .")

    #   BeautifulSoup 
    if page_source:
        print("   BeautifulSoup  ...")
        soup = BeautifulSoup(page_source, 'html.parser')
        print(" .")
    else:
        print("     .")
else:
    print(" 5: WebDriver       <0xEB>.")

#  5   page_source    
# : print(page_source[:1000])
```

```python
# Colab  6 ()
extracted_reviews = [] #    

#   soup    
if soup:
    # print(" 6:  HTML 'comment_text'   <p>     ...") #    
    print(" 6:  HTML 'comment_text'   **     ...") #  

    # soup.find_all('p', class_='comment_text')    
    #  ('p')     
    review_elements = soup.find_all(class_='comment_text') # <----   

    if review_elements:
        print(f" {len(review_elements)} 'comment_text'   .") #   
        for index, element in enumerate(review_elements):
            #       ( )
            tag_name = element.name
            review_text = element.get_text(strip=True)
            if review_text:
                print(f"  -  {index+1} (: <{tag_name}>)    ")
                extracted_reviews.append(review_text)
            # else:
            #    print(f"  -  {index+1} (: <{tag_name}>)    .")

        print(f"  .  {len(extracted_reviews)}     .")
    else:
        #   
        print(":  HTML 'comment_text'      .")
        print("       ,      .")
else:
    print(" 6: soup     <0xEB>.")
```
