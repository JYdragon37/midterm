```python
# VWorld API URL 및 API Key
api_url = "https://api.vworld.kr/req/address?"
api_key = "94194597-6426-3593-8432-EEC2386EDD20"  # 여기에 발급받은 API Key를 입력하세요.
```

```python
import pandas as pd
import requests
import re  # 정규식을 사용해 괄호 안의 내용 제거
from urllib.parse import quote

# 파일 경로 설정
file_path = r"C:\Users\highk\pypy50\27. 전국 대학교 위치 시각화\고등교육기관 하반기 주소록(2023).xlsx"

# Excel 파일 읽기
df = pd.read_excel(file_path, header=5)

# 필요한 열만 추출 (학교명과 주소)
selected_columns = df[['학교명', '주소']]

# 괄호 안의 내용을 제거하는 함수 정의
def clean_address(address):
    return re.sub(r'\([^)]*\)', '', address).strip()  # 괄호와 내부 내용 제거 후 공백 제거

# 주소 정리: 괄호 제거 후 주소 정리
selected_columns['주소'] = selected_columns['주소'].apply(clean_address)

# 상위 5개 데이터 추출
test_data = selected_columns.head(5)

# 결과 저장
coordinates = []

for index, row in test_data.iterrows():
    school_name = row['학교명']
    address = row['주소']
    params = {
        "service": "address",
        "request": "getCoord",
        "key": api_key,  # 외부에서 제공되는 API Key
        "address": quote(address),  # 주소를 URL 인코딩
        "type": "ROAD",  # 도로명 주소
        "format": "json",  # JSON 형식으로 응답
        "crs": "epsg:4326"  # WGS84 좌표계
    }

    # API 요청
    response = requests.get(api_url, params=params)
    print(f"요청 URL: {response.url}")  # 디버깅용 URL 출력

    if response.status_code == 200:
        data = response.json()
        if data['response']['status'] == 'OK':
            # 좌표 추출
            x = data['response']['result']['point']['x']
            y = data['response']['result']['point']['y']
            coordinates.append({"학교명": school_name, "주소": address, "경도(x좌표)": x, "위도(y좌표)": y})
        else:
            coordinates.append({"학교명": school_name, "주소": address, "경도(x좌표)": None, "위도(y좌표)": None})
            print(f"주소 '{address}'에 대한 검색 결과가 없습니다.")
    else:
        print(f"API 요청 실패: {response.status_code}")
        coordinates.append({"학교명": school_name, "주소": address, "경도(x좌표)": None, "위도(y좌표)": None})

# 결과를 Pandas DataFrame으로 변환
df_coordinates = pd.DataFrame(coordinates)

# x좌표와 y좌표만 출력
print(df_coordinates[['학교명', '경도(x좌표)', '위도(y좌표)']])

# CSV로 저장 (옵션)
# df_coordinates.to_csv("대학_주소_좌표_테스트.csv", index=False, encoding="utf-8-sig")
```
